```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('excavation_risk_dataset.csv')
print("Dataset info")
df.info()
```

```python
print("\nMissing Values")
print(df.isnull().sum())
```

```python
print("\nSummary Statistics")
print(df.describe())
```

```python
print("\nValue Counts for Categorical Columns")
cat_cols =df.select_dtypes(include=['object']).columns
for col in cat_cols:
    print(f"\n{col}:")
    print(df[col].value_counts())
```

```python
sns.set_theme(style="whitegrid")
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='Risk_Level', hue='Risk_Level', palette='viridis', legend=False)
plt.title('Distribution of Risk Level')
plt.tight_layout()
plt.savefig('risk_level_dist.png')
plt.close()
```

```python
plt.figure(figsize=(12, 10))
corr = df.select_dtypes(include=['float64', 'int64']).corr()
sns.heatmap(corr, annot=False, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.savefig('correlation_matrix.png')
plt.close()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='Risk_Level', y='Deformation_mm', hue='Risk_Level', palette='Set2', legend=False)
plt.title('Deformation vs Risk Level')
plt.tight_layout()
plt.savefig('deformation_vs_risk.png')
plt.close()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='Risk_Level', y='Ground_Settlement_mm', hue='Risk_Level', palette='Set1', legend=False)
plt.title('Ground Settlement vs Risk Level')
plt.tight_layout()
plt.savefig('settlement_vs_risk.png')
plt.close()
```
